```python
import pandas as pd
```

```python
df_mxp = pd.read_csv('/Users/adamzsolt/Documents/CCT/CA2/FAOSTAT_data_5-4-2022-8.csv')
#https://www.fao.org/faostat/en/#data/TI
df_mxp
```

```python
df_mxp.shape
```

```python
df_mxp.isnull().sum()
```

```python
df_mxp = df_mxp.dropna()
```

```python
df_mxp.isnull().sum()
```

```python
df_mxp.shape
```

```python
df_mxp = df_mxp[['Unnamed: 0','Unnamed: 1','Unnamed: 2',
                 'Export Value Base Price [95].1','Import Value Base Period Price [65].1',
                 'Export Value Index (2014-2016 = 100) [492].1','Import Value Index (2014-2016 = 100) [462].1']]

df_mxp = df_mxp.rename(columns={'Unnamed: 0':'Year','Unnamed: 1':'Item','Unnamed: 2':'Country',
                                'Export Value Index (2014-2016 = 100) [492].1':'XPI','Export Value Base Price [95].1':'Export_value',
                                'Import Value Index (2014-2016 = 100) [462].1':'MPI','Import Value Base Period Price [65].1':'Import_value'})

df_mxp.head()
```

```python
df_mxp['Year'] = df_mxp['Year'].apply(lambda x: x[:4])
df_mxp['Item'] = df_mxp['Item'].apply(lambda x: x.split()[1])
df_mxp['Country'] = df_mxp['Country'].apply(lambda x: x.split()[0])
df_mxp.head()
```

```python
df_mxp = df_mxp.sort_values(['Country','Year'], ascending=[True, True])
#using sort function I got the values in order by the year and the item name
df_mxp.head()
```

```python
df_mxp.shape
```

```python
df_ppi = pd.read_csv('/Users/adamzsolt/Documents/CCT/CA2/FAOSTAT_data_5-5-2022.csv')
#https://www.fao.org/faostat/en/#data/PP
df_ppi.head()
```

```python
df_ppi = df_ppi[['Value','Item','Year','Area']]
df_ppi['Item'] = df_ppi['Item'].apply(lambda x: x.split()[1])
df_ppi = df_ppi.rename(columns={'Value':'PPI','Area':'Country'})
df_ppi.head()
```

```python
df_ppi.shape,df_mxp.shape
```

```python
df_mxp['Year'] = df_mxp['Year'].astype(int)
df_ppi['Year'] = df_ppi['Year'].astype(int)
df = pd.merge(df_mxp,df_ppi, on=['Year','Item','Country'], how='right')
```
